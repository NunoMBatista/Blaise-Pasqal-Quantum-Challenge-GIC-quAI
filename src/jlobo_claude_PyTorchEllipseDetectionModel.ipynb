```python
# Ellipse Detection in 8x8 Grayscale Images with PyTorch
# Colab Notebook

import numpy as np
import matplotlib.pyplot as plt
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image, ImageDraw
from google.colab import drive
from tqdm.notebook import tqdm
import random
import time

# Mount Google Drive (if using Colab)
try:
    drive.mount('/content/drive')
    print("Google Drive mounted successfully")
except:
    print("Running locally")

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)
random.seed(42)

# Configuration parameters
IMAGE_SIZE = 8
GRAYSCALE_LEVELS = 8
NUM_IMAGES = 256
BATCH_SIZE = 32
LEARNING_RATE = 0.001
EPOCHS = 50
DATASET_DIR = "pixel_art_dataset"

# Check if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Part 1: Generate Dataset with Labels
# ===================================

def create_smooth_gradient_background(size):
    """Create a smooth random gradient background."""
    # Generate four corner values
    corners = np.random.randint(0, GRAYSCALE_LEVELS, size=4) * (255 // (GRAYSCALE_LEVELS - 1))

    # Create x and y grids
    y, x = np.mgrid[0:size, 0:size]
    y = y / (size - 1)
    x = x / (size - 1)

    # Bilinear interpolation
    top = corners[0] * (1 - x) + corners[1] * x
    bottom = corners[2] * (1 - x) + corners[3] * x
    gradient = top * (1 - y) + bottom * y

    # Quantize to specified grayscale levels
    gradient = np.round(gradient / (255 // (GRAYSCALE_LEVELS - 1))) * (255 // (GRAYSCALE_LEVELS - 1))

    return gradient.astype(np.uint8)

def add_ellipse(img_array, size):
    """Add an ellipse with defined contours and uniform fill. Returns the modified array and ellipse parameters."""
    # Create a temporary array for the ellipse mask
    ellipse_mask = np.zeros((size, size), dtype=np.uint8)

    # Create a PIL image to draw on
    temp_img = Image.fromarray(ellipse_mask, 'L')
    draw = ImageDraw.Draw(temp_img)

    # Randomly determine ellipse properties
    width = random.randint(2, size-1)
    height = random.randint(2, size-1)
    x0 = random.randint(0, size - width)
    y0 = random.randint(0, size - height)
    x1 = x0 + width
    y1 = y0 + height

    # Calculate center and radius for bounding box format
    center_x = (x0 + x1) / 2
    center_y = (y0 + y1) / 2
    radius_x = width / 2
    radius_y = height / 2

    # Choose a random grayscale level for the fill
    fill_value = random.randint(0, GRAYSCALE_LEVELS-1) * (255 // (GRAYSCALE_LEVELS - 1))

    # Draw ellipse
    draw.ellipse([x0, y0, x1, y1], fill=255)  # Use 255 for the mask

    # Convert back to array
    ellipse_mask = np.array(temp_img) > 0

    # Apply the ellipse to the original image
    img_array[ellipse_mask] = fill_value

    # Return the modified image and the ellipse parameters
    ellipse_params = {
        'center_x': center_x / size,  # Normalize to [0,1]
        'center_y': center_y / size,  # Normalize to [0,1]
        'radius_x': radius_x / size,  # Normalize to [0,1]
        'radius_y': radius_y / size,  # Normalize to [0,1]
        'present': 1.0  # Indicator that an ellipse is present (for multi-label classification)
    }

    return img_array, ellipse_params, ellipse_mask

def generate_dataset():
    """Generate dataset with labels indicating whether images contain ellipses and their parameters."""
    if not os.path.exists(DATASET_DIR):
        os.makedirs(DATASET_DIR)

    images = []
    labels = []
    masks = []

    print("Generating dataset...")
    for i in tqdm(range(NUM_IMAGES)):
        # Create a smooth gradient background
        img_array = create_smooth_gradient_background(IMAGE_SIZE)

        # Default labels for no ellipse
        ellipse_params = {
            'center_x': 0.0,
            'center_y': 0.0,
            'radius_x': 0.0,
            'radius_y': 0.0,
            'present': 0.0
        }

        ellipse_mask = np.zeros((IMAGE_SIZE, IMAGE_SIZE), dtype=bool)

        # Decide if we should add ellipses (50% chance)
        if random.random() < 0.5:
            # Add just one ellipse for simplicity in this training example
            img_array, ellipse_params, ellipse_mask = add_ellipse(img_array, IMAGE_SIZE)

        # Create PIL image from array and save
        img = Image.fromarray(img_array, 'L')
        img.save(os.path.join(DATASET_DIR, f"pixel_art_{i:03d}.jpg"), quality=100)

        # Store image and labels
        images.append(img_array)
        labels.append(ellipse_params)
        masks.append(ellipse_mask)

    return images, labels, masks

# If dataset doesn't exist, generate it
if not os.path.exists(DATASET_DIR) or len(os.listdir(DATASET_DIR)) < NUM_IMAGES:
    images, labels, masks = generate_dataset()
    # Save labels for later use
    np.save(os.path.join(DATASET_DIR, "labels.npy"), labels)
    print(f"Dataset created with {len(images)} images")
else:
    print(f"Dataset already exists in {DATASET_DIR}")
    # Try to load labels if they exist
    try:
        labels = np.load(os.path.join(DATASET_DIR, "labels.npy"), allow_pickle=True)
        print(f"Labels loaded with {len(labels)} entries")
    except:
        print("Labels not found. Please regenerate the dataset.")

# Part 2: Custom Dataset Class
# ===========================

class EllipseDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.image_files = [f for f in os.listdir(root_dir) if f.endswith('.jpg')]
        self.image_files.sort()

        # Load labels
        labels_path = os.path.join(root_dir, "labels.npy")
        if os.path.exists(labels_path):
            self.labels = np.load(labels_path, allow_pickle=True)
        else:
            raise FileNotFoundError(f"Labels file not found at {labels_path}")

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_path = os.path.join(self.root_dir, self.image_files[idx])
        image = Image.open(img_path).convert('L')

        if self.transform:
            image = self.transform(image)

        # Extract label components
        label = self.labels[idx]

        # Create target tensor: [present, center_x, center_y, radius_x, radius_y]
        target = torch.tensor([
            label['present'],
            label['center_x'],
            label['center_y'],
            label['radius_x'],
            label['radius_y']
        ], dtype=torch.float32)

        return image, target

# Define transforms
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Create train/validation split
dataset = EllipseDataset(root_dir=DATASET_DIR, transform=transform)
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(dataset, [train_size, val_size])

# Create data loaders
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)

print(f"Training data: {len(train_dataset)} samples")
print(f"Validation data: {len(val_dataset)} samples")

# Part 3: Visualization functions
# ==============================

def visualize_sample(image, target, prediction=None, title="Sample Image with Ellipse"):
    """Visualize an image with its ground truth ellipse and optionally the predicted ellipse."""
    # Convert image tensor to numpy array
    if torch.is_tensor(image):
        img = image.cpu().numpy().squeeze()
        # Denormalize
        img = (img * 0.5 + 0.5) * 255
    else:
        img = image

    # Create figure
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.imshow(img, cmap='gray', vmin=0, vmax=255)
    ax.set_title(title)

    # Extract target parameters
    is_present = target[0].item() > 0.5

    if is_present:
        center_x = target[1].item() * IMAGE_SIZE
        center_y = target[2].item() * IMAGE_SIZE
        radius_x = target[3].item() * IMAGE_SIZE
        radius_y = target[4].item() * IMAGE_SIZE

        # Add ellipse to plot
        from matplotlib.patches import Ellipse
        ellipse = Ellipse((center_x, center_y), 2*radius_x, 2*radius_y,
                          fill=False, edgecolor='green', linewidth=2)
        ax.add_patch(ellipse)
        ax.text(0, -0.5, "Ground truth", color='green', transform=ax.transAxes)

    # Add prediction if provided
    if prediction is not None:
        pred_present = prediction[0].item() > 0.5

        if pred_present:
            pred_center_x = prediction[1].item() * IMAGE_SIZE
            pred_center_y = prediction[2].item() * IMAGE_SIZE
            pred_radius_x = prediction[3].item() * IMAGE_SIZE
            pred_radius_y = prediction[4].item() * IMAGE_SIZE

            # Add predicted ellipse to plot
            from matplotlib.patches import Ellipse
            pred_ellipse = Ellipse((pred_center_x, pred_center_y), 2*pred_radius_x, 2*pred_radius_y,
                                   fill=False, edgecolor='red', linewidth=2, linestyle='--')
            ax.add_patch(pred_ellipse)
            ax.text(0.7, -0.5, "Prediction", color='red', transform=ax.transAxes)

    plt.tight_layout()
    plt.show()

# Visualize some samples
def visualize_samples(loader, num_samples=5):
    # Get some random samples
    dataiter = iter(loader)
    images, targets = next(dataiter)

    for i in range(min(num_samples, len(images))):
        visualize_sample(images[i], targets[i])

print("Visualizing some training samples:")
visualize_samples(train_loader)

# Part 4: Model Definition
# =======================

class EllipseDetector(nn.Module):
    def __init__(self):
        super(EllipseDetector, self).__init__()

        # Convolutional layers
        self.features = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)  # Output: 32 x 4 x 4
        )

        # Fully connected layers
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(32 * 4 * 4, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 5)  # 5 outputs: [present, center_x, center_y, radius_x, radius_y]
        )

        # Activation functions for different outputs
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)

        # Apply sigmoid to all outputs for normalization
        x = self.sigmoid(x)

        return x

# Instantiate the model
model = EllipseDetector().to(device)
print(model)

# Part 5: Loss Function and Optimizer
# ==================================

# Custom loss function: Binary Cross Entropy for presence, MSE for coordinates
def ellipse_loss(predictions, targets):
    # Extract components
    pred_present = predictions[:, 0]
    target_present = targets[:, 0]

    # BCE loss for ellipse presence
    presence_loss = nn.BCELoss()(pred_present, target_present)

    # Only compute coordinate loss for samples where ellipse is actually present
    mask = (target_present > 0.5).float()
    if mask.sum() > 0:
        # MSE loss for coordinates (normalized by number of positive samples)
        pred_coords = predictions[:, 1:]
        target_coords = targets[:, 1:]

        # Apply mask to focus on samples with ellipses
        coord_loss = (nn.MSELoss(reduction='none')(pred_coords, target_coords) * mask.unsqueeze(1)).sum() / (mask.sum() + 1e-8)
    else:
        coord_loss = torch.tensor(0.0).to(device)

    # Total loss
    total_loss = presence_loss + 10.0 * coord_loss  # Weighting factor to balance losses

    return total_loss, presence_loss, coord_loss

# Initialize optimizer
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

# Part 6: Training Loop
# ====================

def train_model(model, train_loader, val_loader, optimizer, num_epochs):
    # Training history
    history = {
        'train_loss': [],
        'val_loss': [],
        'train_presence_acc': [],
        'val_presence_acc': []
    }

    best_val_loss = float('inf')

    for epoch in range(num_epochs):
        start_time = time.time()

        # Training phase
        model.train()
        train_loss = 0.0
        train_presence_correct = 0
        train_samples = 0

        for images, targets in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Train]"):
            images = images.to(device)
            targets = targets.to(device)

            # Zero the gradients
            optimizer.zero_grad()

            # Forward pass
            outputs = model(images)

            # Compute loss
            loss, presence_loss, coord_loss = ellipse_loss(outputs, targets)

            # Backward pass and optimize
            loss.backward()
            optimizer.step()

            # Update statistics
            train_loss += loss.item() * images.size(0)

            # Calculate presence accuracy
            pred_present = (outputs[:, 0] > 0.5).float()
            target_present = (targets[:, 0] > 0.5).float()
            train_presence_correct += (pred_present == target_present).sum().item()
            train_samples += images.size(0)

        avg_train_loss = train_loss / train_samples
        train_presence_acc = train_presence_correct / train_samples

        # Validation phase
        model.eval()
        val_loss = 0.0
        val_presence_correct = 0
        val_samples = 0

        with torch.no_grad():
            for images, targets in tqdm(val_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Val]"):
                images = images.to(device)
                targets = targets.to(device)

                # Forward pass
                outputs = model(images)

                # Compute loss
                loss, presence_loss, coord_loss = ellipse_loss(outputs, targets)

                # Update statistics
                val_loss += loss.item() * images.size(0)

                # Calculate presence accuracy
                pred_present = (outputs[:, 0] > 0.5).float()
                target_present = (targets[:, 0] > 0.5).float()
                val_presence_correct += (pred_present == target_present).sum().item()
                val_samples += images.size(0)

        avg_val_loss = val_loss / val_samples
        val_presence_acc = val_presence_correct / val_samples

        # Update history
        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['train_presence_acc'].append(train_presence_acc)
        history['val_presence_acc'].append(val_presence_acc)

        # Save best model
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            torch.save(model.state_dict(), 'best_ellipse_detector.pth')
            print(f"Saved new best model with validation loss: {best_val_loss:.4f}")

        # Print epoch statistics
        time_elapsed = time.time() - start_time
        print(f"Epoch {epoch+1}/{num_epochs} completed in {time_elapsed:.0f}s")
        print(f"Train Loss: {avg_train_loss:.4f}, Train Presence Acc: {train_presence_acc:.4f}")
        print(f"Val Loss: {avg_val_loss:.4f}, Val Presence Acc: {val_presence_acc:.4f}")
        print("-" * 60)

    # Load best model
    model.load_state_dict(torch.load('best_ellipse_detector.pth'))

    return model, history

# Train the model
model, history = train_model(model, train_loader, val_loader, optimizer, num_epochs=EPOCHS)

# Part 7: Visualize Results
# ========================

# Plot training history
plt.figure(figsize=(12, 5))

# Plot loss
plt.subplot(1, 2, 1)
plt.plot(history['train_loss'], label='Train Loss')
plt.plot(history['val_loss'], label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()

# Plot accuracy
plt.subplot(1, 2, 2)
plt.plot(history['train_presence_acc'], label='Train Acc')
plt.plot(history['val_presence_acc'], label='Val Acc')
plt.xlabel('Epoch')
plt.ylabel('Presence Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()

plt.tight_layout()
plt.show()

# Part 8: Evaluate on Test Samples
# ===============================

def evaluate_model(model, loader, num_samples=10):
    model.eval()

    # Get some random samples
    all_images = []
    all_targets = []
    all_preds = []

    with torch.no_grad():
        for images, targets in loader:
            images = images.to(device)
            outputs = model(images)

            all_images.extend(images.cpu())
            all_targets.extend(targets)
            all_preds.extend(outputs.cpu())

            if len(all_images) >= num_samples:
                break

    # Visualize predictions
    for i in range(min(num_samples, len(all_images))):
        visualize_sample(all_images[i], all_targets[i], all_preds[i],
                        title=f"Prediction (Present: {all_preds[i][0].item():.2f})")

print("\nEvaluating model on validation samples:")
evaluate_model(model, val_loader)

# Part 9: Model Inference on New Images
# ====================================

def predict_ellipses(model, image_path):
    """Predict ellipses in a new image."""
    # Load and preprocess image
    image = Image.open(image_path).convert('L')
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))
    ])
    image_tensor = transform(image).unsqueeze(0).to(device)

    # Make prediction
    model.eval()
    with torch.no_grad():
        prediction = model(image_tensor)

    # Get numpy arrays
    img_np = image
    pred_np = prediction[0].cpu().numpy()

    # Create synthetic target (all zeros) just for visualization
    target = torch.zeros(5)
    target[0] = pred_np[0]  # Set presence same as prediction
    target[1:] = torch.tensor(pred_np[1:])  # Set coordinates same as prediction

    # Visualize
    visualize_sample(np.array(img_np), target, torch.tensor(pred_np),
                    title=f"Inference Result (Confidence: {pred_np[0]:.2f})")

    return pred_np

# Example usage:
# predict_ellipses(model, os.path.join(DATASET_DIR, "pixel_art_000.jpg"))

# Test on a few random images
import random
image_files = [f for f in os.listdir(DATASET_DIR) if f.endswith('.jpg')]
random_images = random.sample(image_files, 5)

print("\nTesting model on random images:")
for img_file in random_images:
    img_path = os.path.join(DATASET_DIR, img_file)
    print(f"\nPredicting ellipse in {img_file}")
    prediction = predict_ellipses(model, img_path)
    print(f"Prediction: Present={prediction[0]:.2f}, Center=({prediction[1]:.2f}, {prediction[2]:.2f}), Radius=({prediction[3]:.2f}, {prediction[4]:.2f})")

# Part 10: Evaluate Model Performance
# ==================================

def compute_metrics(model, loader):
    """Compute detailed metrics for the model."""
    model.eval()

    all_preds = []
    all_targets = []

    with torch.no_grad():
        for images, targets in tqdm(loader, desc="Computing metrics"):
            images = images.to(device)
            outputs = model(images)

            all_preds.extend(outputs.cpu().numpy())
            all_targets.extend(targets.numpy())

    all_preds = np.array(all_preds)
    all_targets = np.array(all_targets)

    # Compute presence metrics
    pred_present = (all_preds[:, 0] > 0.5).astype(int)
    actual_present = (all_targets[:, 0] > 0.5).astype(int)

    presence_accuracy = (pred_present == actual_present).mean()

    true_positives = ((pred_present == 1) & (actual_present == 1)).sum()
    false_positives = ((pred_present == 1) & (actual_present == 0)).sum()
    false_negatives = ((pred_present == 0) & (actual_present == 1)).sum()

    precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
    f1_score = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0

    # Compute localization metrics for true positives only
    if true_positives > 0:
        # Filter for images with ellipses
        mask = (actual_present == 1)
        pred_coords = all_preds[mask, 1:]
        true_coords = all_targets[mask, 1:]

        # Mean squared error for coordinates
        coord_mse = ((pred_coords - true_coords) ** 2).mean()

        # IoU (Intersection over Union) for ellipses - approximate using centers and radii
        ious = []
        for i in range(len(pred_coords)):
            pred_center_x, pred_center_y, pred_radius_x, pred_radius_y = pred_coords[i]
            true_center_x, true_center_y, true_radius_x, true_radius_y = true_coords[i]

            # Calculate center distance
            center_dist = np.sqrt((pred_center_x - true_center_x)**2 + (pred_center_y - true_center_y)**2)

            # Simple approximation of ellipse IoU - not perfect but gives an idea
            if center_dist < (pred_radius_x + true_radius_x) and center_dist < (pred_radius_y + true_radius_y):
                # Area of predicted ellipse:  * a * b
                pred_area = np.pi * pred_radius_x * pred_radius_y

                # Area of true ellipse:  * a * b
                true_area = np.pi * true_radius_x * true_radius_y

                # Approximate intersection (very rough approximation)
                overlap_ratio = max(0, 1 - center_dist / ((pred_radius_x + true_radius_x + pred_radius_y + true_radius_y) / 2))
                intersection = overlap_ratio * min(pred_area, true_area)

                # Calculate IoU
                union = pred_area + true_area - intersection
                iou = intersection / union
                ious.append(iou)
            else:
                ious.append(0)  # No overlap

        mean_iou = np.mean(ious)
    else:
        coord_mse = float('nan')
        mean_iou = float('nan')

    # Print results
    print("\nModel Performance Metrics:")
    print(f"Presence Accuracy: {presence_accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1_score:.4f}")

    if not np.isnan(coord_mse):
        print(f"Coordinate MSE: {coord_mse:.4f}")
        print(f"Mean IoU: {mean_iou:.4f}")
    else:
        print("No true positives found. Cannot compute coordinate metrics.")

    return {
        'presence_accuracy': presence_accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1_score,
        'coord_mse': coord_mse,
        'mean_iou': mean_iou
    }

# Compute metrics on validation set
print("\nComputing final model metrics:")
metrics = compute_metrics(model, val_loader)

# Part 11: Save and Load Model
# ===========================

# Save model
torch.save({
    'model_state_dict': model.state_dict(),
    'optimizer_state_dict': optimizer.state_dict(),
    'metrics': metrics
}, 'ellipse_detector_full.pth')

print("\nModel saved to 'ellipse_detector_full.pth'")

# Function to load model
def load_model():
    model = EllipseDetector().to(device)
    checkpoint = torch.load('ellipse_detector_full.pth')
    model.load_state_dict(checkpoint['model_state_dict'])
    model.eval()
    return model

# Example usage to load the model
# loaded_model = load_model()

print("\nTraining and evaluation complete!")
```
